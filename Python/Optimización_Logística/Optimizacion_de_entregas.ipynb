```python
import pandas as pd

# Datos de entregas del día
entregas = {
    'Ruta': ['Cali-Norte', 'Cali-Sur', 'Yumbo-Industrial', 'Jamundí-Centro', 'Cali-Oeste'],
    'Tiempo_Estimado_Min': [30, 45, 60, 40, 35],
    'Tiempo_Real_Min': [55, 40, 110, 35, 85],
    'Costo_Envio': [50000, 45000, 80000, 38000, 42000]
}

df_logistica = pd.DataFrame(entregas)
```

```python
df_logistica
```

```python
# 1. Calcular la 'Desviación'
df_logistica['Desviacion'] = df_logistica['Tiempo_Real_Min'] - df_logistica['Tiempo_Estimado_Min']
df_logistica
```

```python
# 2. Crear la columna 'Penalización'
df_logistica['Penalizacion'] = df_logistica.apply(lambda x: x['Costo_Envio'] * 0.5 if x['Desviacion'] > 45 else (x['Costo_Envio'] * 0.20 if x['Desviacion'] > 20 else 0 ), axis=1)
df_logistica
```

```python
# 3. Calcular el 'Ingreso_Neto'
df_logistica['Ingreso_Neto'] = df_logistica['Costo_Envio'] - df_logistica['Penalizacion']
df_logistica
```
