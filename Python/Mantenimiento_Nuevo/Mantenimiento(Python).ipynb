```python
df_mantenimiento['Horas_Parada'] = df_mantenimiento['Horas_Parada'].astype(str).str.replace(',', '.')
df_mantenimiento
```

```python
mediana = df_mantenimiento['Horas_Parada'].median()
df_mantenimiento['Horas_Parada'].fillna(mediana)
df_mantenimiento
```

```python
df_mantenimiento['Horas_Parada'] = pd.to_numeric(df_mantenimiento['Horas_Parada'], errors='coerce')
df_mantenimiento.info()
```

```python
df_mantenimiento
```

```python
mediana = df_mantenimiento['Horas_Parada'].median()
df_mantenimiento['Horas_Parada'] = df_mantenimiento['Horas_Parada'].fillna(mediana)
df_mantenimiento
```

```python
df_mantenimiento['Costo_Hora'] = df_mantenimiento['Costo_Reparacion'] / df_mantenimiento['Horas_Parada']
df_mantenimiento['Costo_Hora'] = df_mantenimiento['Costo_Hora'].round(2)
df_mantenimiento
```

```python
Agrupar = df_mantenimiento.groupby('Maquina')['Costo_Hora'].sum()
print(Agrupar)
```

```python
Agrupa_Cosot = df_mantenimiento.groupby('Maquina')['Costo_Reparacion'].sum()
print(Agrupa_Cosot)
```

```python
Agrupaar_Horas = df_mantenimiento.groupby('Maquina')['Horas_Parada'].sum()
print(Agrupaar_Horas)
```
